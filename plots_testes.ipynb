```python
import pandas as pd
```

```python
df = pd.read_csv('valores_Infectados.csv', index_col=[0])
```

```python
df.head()
```

```python
df.plot().legend(bbox_to_anchor=(1,1)) #verificar o porque do 'Unnamed'...
```

```python
import matplotlib.pyplot as plt

colors = []

def set_colors(df):
    for col in df.columns:
        if 'Icum1' in col:
            colors.append('#990000')
        else:
            colors.append('#999999')
    return df
```

```python
fig, ax = plt.subplots(dpi=600, figsize=(10,4))

(df
 .pipe(set_colors)
 .plot(ax=ax, rot=45, color=colors)
 .legend(bbox_to_anchor=(1,1))
 )
```

```python
fig, ax = plt.subplots(dpi=600, figsize=(10,4))

(df
 .sum()
 .plot.bar(rot=45, figsize=(12,4), ax=ax)
 )
ax.set_xticklabels(labels=ax.get_xticklabels(), ha='right')
ax.legend(bbox_to_anchor=(1,1))
```
